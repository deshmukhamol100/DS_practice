# 2.Array Manipulation function

```python
#reshape the array
a1 = np.array([1,2,3])
reshaped_a1 = np.reshape(a1,(3,1))#reshape the array a1 to have 3 rows and 1 column
reshaped_a1
```

```python
f1 = np.array([[1,2,3],[4,5,6]])
falternate_f1 = np.ravel(f1) # ravel the array f1 to a 1D array
falternate_f1
```

```python
e1 = np.array([[1,2,3],[4,5,6]])
tranposed_e1 = np.transpose(e1) # transpose the array e1
tranposed_e1
```

```python
a1 = np.array([1,2,3])
a2 =np.array([4,5,6 ])
stacked_a2 = np.vstack((a1,a2)) # stack the array a2 on top of itself
stacked_a2
```

```python
a1 = np.array([1,2,3])
a2 =np.array([4,5,6 ])
stacked_a2 = np.hstack((a1,a2)) # stack the array a2 on top of itself
stacked_a2
```
