# If elif else 

```python
x= 5
r = x % 2
if r == 0:
    print("x is even.")
elif x > 5:
    print("x is odd and greater than 5.")
elif x < 5:
    print("x is odd and less than 5.")
else:
    print("x is odd and not greater than 5.")
```

<h4> Increament pattern

```python
for i in range(4):
    for j in range(i+1):
        print("*", end=' ')
    print()
```

```python
for i in range(10):
    for j in range(i+1):
        print("*", end='  ')
    print() 
```

```python
for i in range(4,0,-1):
    print("* " * i)
```

```python
for i in range(7,0,-1):
   print(''*(5*i) + "* " *(1*i))
```
